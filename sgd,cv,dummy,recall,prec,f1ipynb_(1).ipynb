multiclass classificatin

OvA or OvR  one versus all or one versus rest en yaxsi scoru take

Ovo one vers one  compare all w each other  ve her biri ucun 1 binary classifi duzeldir  10 class 45 pairs 10*9/2

SVC (Support Vector Classifier) in scikit-learn always uses OvO (One-vs-One) for multiclass problems.

LinearSVC (a faster linear version) uses OvA (One-vs-All).

```python
from sklearn.svm import SVC

svm_clf =SVC(random_state = 42 )
svm_clf.fit(X_train[:2000],y_train[:2000])
```

```python
svm_clf.predict([some_digit])
```

```python
some_digit_scores =svm_clf.decision_function([some_digit])
some_digit_scores.round(2) # The machine doesn’t just say “YES/NO.” It gives a confidence number for each possible digit (0–9).
```

```python
class_id = some_digit_scores.argmax()
class_id #It looks through the array and finds the position of the biggest number
```

```python
svm_clf.classes_ #It lists all the unique class labels the model has learned during training.
#f you trained on MNIST digits (0–9) will output:
```

```python
svm_clf.classes_[class_id]
```

```python
svm_clf.decision_function_shape='ovo'
some_digit_scores_ovo =svm_clf.decision_function([some_digit])
some_digit_scores_ovo.round(2)
```
